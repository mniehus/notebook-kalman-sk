# Súčet dvojrozmerných normálnych rozdelení

Pre súčet dvojrozmerných funkcií je možné odvodiť podobne ako pre súčin - [link](http://www.tina-vision.net/docs/memos/2003-003.pdf), podobne ako pri jednorozmernom prípade, vzťahy pre strednú hodnotu a rozptyl

\begin{equation}
\nonumber
\mu = \mu_1 + \mu_2
\end{equation}

\begin{equation}
\nonumber
\Sigma = \Sigma_1 + \Sigma_2
\end{equation}

Pre výpočet a zobrazenie súčtu pdf použijeme rovnaký program ako v prípade súčinu s malými úpravami.


```python
%reset -f  
%matplotlib inline

# Priklad - Sucet dvoch normalnych rozdeleni

from numpy import array, dot
from numpy.linalg import inv

m1=array([0,0])
S1=array([[5, 2.5], [2.5, 2]])

m2=array([1,1])
S2=array([[6, -4], [-4, 3]])

m3 = m1 + m2
S3 = S1 + S2

print('Mean \n',m3)
print()
print('Sigma\n',S3)
```

```python
from numpy.random import multivariate_normal
import matplotlib.pylab as plt
from src.utils import *

# Priklad - Zobrazenie suctu dvoch normalnych rozdeleni pomocou elipse kovariancie.

fig = plt.figure(0)

X1, Y1 = multivariate_normal(m1, S1, 1000).T
plt.plot(X1, Y1, 'r.', alpha=.2, label='m1, S1')
cov_ellipse(fig, m1,S1, ec='r', fc='r', alpha = 0.7, fill = False)

X2, Y2 = multivariate_normal(m2, S2, 1000).T
plt.plot(X2, Y2, 'g.', alpha=.2, label='m2, S2')
cov_ellipse(fig, m2,S2, ec='g', fc='g', alpha = 0.7, fill = False)

X3, Y3 = multivariate_normal(m3, S3, 250).T
cov_ellipse(fig, m3,S3, ec='b', fc='b', alpha=0.7, fill=False)
plt.plot(X3, Y3, 'b.', alpha=.2, label='m3, S3')

plt.ylim(-10,10)
plt.axis('equal')
plt.legend()
plt.grid()
plt.show()
```

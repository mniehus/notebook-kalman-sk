# Jednorozmerná spojitá náhodná premenná


V prípade spojitého rozloženia hodnôt $x$ môžeme vzťahy pre strednú hodnotu a rozptyl upraviť limitným prechodom vzťahov pre diskrétne rozdelenie do interálneho tvaru.

Pre strednú hodnotu spojitého rozdelenia platí 
\begin{equation}
E[X] =  \int_{-\infty}^{\infty} p(x) \, x \,  dx 
\end{equation}

kde $p(x)$ je funkcia rozdelenia hustoty pravdepodobnosti (spojitej) náhodnej premennej $X$ (<i>probability density function - pdf </i>), pre ktorú podobne ako pri diskrétnom rozdelení platí
    
\begin{equation}
\int_{-\infty}^{\infty} p(x) \, dx = 1 
\end{equation}

Vzťah pre rozptyl má tvar  

\begin{equation}
Var[X]  = E[(X-\mu)^2] = \int_{-\infty}^{\infty} (x - \mu)^2 \, p(x) \, dx 
\end{equation}

Štandardná odchýlka (<i>standard deviation</i>) $\sigma$ je definovaná ako

\begin{equation}
\sigma = \sqrt{Var[X]} 
\end{equation}

Pre náhodnú premennú môžeme na základe funkcie hustoty pravdepodobnosti <i>pdf</i> definovať funkciu rozdelenia pravdepodobnosti $P(x)$ (<i>distribučná funkcia, probability distribution function - PDF, cumulative distribution finction - CDF </i>) ako

\begin{equation}
P(x) = P(X < x) = \int_{-\infty}^x p(z) \, dz 
\end{equation}



Pre hustotu pravdepodobnosti $p(x)$ a distribučnú funkciu $P(x)$ platia nasledujúce vzťahy: 

\begin{equation}
p(x) \geq 0
\end{equation}



\begin{equation}
 \int_{-\infty}^{\infty} p(x) \, dx = 1
\end{equation}

\begin{equation} 
p(x) = \frac{d \, P(x)}{d \, x} 
\end{equation}

\begin{equation} 
P(a \leq X \leq b) = \int_a^b p(x) dx 
\end{equation}

\begin{equation}
 P(-\infty) = 0
\end{equation}

\begin{equation}
P(\infty) = 1 
\end{equation}

\begin{equation}
P(x) \in [0, 1] 
\end{equation}

\begin{equation}
P(a) \leq P(b), \quad a \leq b 
\end{equation}

\begin{equation}
P(a \leq X \leq b) = P(b) - P(a)
\end{equation}

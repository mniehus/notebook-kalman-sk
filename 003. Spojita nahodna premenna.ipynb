# Jednorozmerná spojitá náhodná premenná


V prípade ak hodnoty $x_i$ náhodnej premennej X majú spojité rozloženie, môžeme vzťahy pre strednú hodnotu a rozptyl získať úpravou vzťahov pre diskrétne rozdelenie. Z množiny hodnôt pravdepodobnosti $P(x_i)$ pre diskrétne hodnoty $x_i$ ziskame funkciu rozdelenia hustoty pravdepodobnosti $f(x)$ prechodom k spojitému rozdeleniu hodnôt náhodnej premennej

\begin{equation}
\{ P(x_0), \, P(x_1), \dots   P(x_n)  \}  = P(x_i) \quad i= 0 \dots n 
\end{equation}


\begin{equation}
pre \quad n \rightarrow \pm \infty \quad P(x_i) \rightarrow f(x)
\end{equation}

Funkcia $f(x)$ býva v literatúre označovaná aj ako pdf - probability density function. Pre funkciu $f(x)$ podobne ako pri diskrétnom rozdelení platí
    
\begin{equation}
\int_{-\infty}^{\infty} f(x) \, dx = 1 
\end{equation}

Pre strednú hodnotu spojitého rozdelenia platí 

\begin{equation}
E[X] =  \int_{-\infty}^{\infty} f(x) \, x \,  dx 
\end{equation}

Podobne ako pri diskrétnom rozdelení môžeme $E[X]$ pokladať za operátor nad premennou X, napr.

\begin{equation}
E[X-a] =  \int_{-\infty}^{\infty} f(x) \, (x-a) \,  dx 
\end{equation}


Vzťah pre rozptyl má tvar  

\begin{equation}
Var[X]  = E[(X-\mu)^2] = \int_{-\infty}^{\infty} f(x) \,(x - \mu)^2 \,  dx 
\end{equation}

Štandardná odchýlka (<i>standard deviation</i>) $\sigma$ je definovaná ako

\begin{equation}
\sigma = \sqrt{Var[X]} 
\end{equation}

Pre náhodnú premennú môžeme na základe funkcie hustoty pravdepodobnosti <i>f(x)</i> definovať funkciu rozdelenia pravdepodobnosti $F(x)$ ako

\begin{equation}
F(x) = P(X < x) = \int_{-\infty}^x f(z) \, dz 
\end{equation}

V literatúre býva funkcia rozdelenia pravdepodobnosti $F(x)$ označovaná aj ako distribučná funkcia, PDF - probability distribution function alebo CDF - cumulative distribution function.



Pre hustotu pravdepodobnosti $f(x)$ a distribučnú funkciu $F(x)$ platia nasledujúce vzťahy: 

\begin{equation}
f(x) \geq 0
\end{equation}

\begin{equation}
 \int_{-\infty}^{\infty} f(x) \, dx = 1
\end{equation}

\begin{equation} 
f(x) = \frac{d \, F(x)}{d \, x} 
\end{equation}

\begin{equation} 
P(a \leq x \leq b) = \int_a^b f(x) dx 
\end{equation}

\begin{equation}
 F(-\infty) = 0
\end{equation}

\begin{equation}
F(\infty) = 1 
\end{equation}

\begin{equation}
F(x) \in [0, 1] 
\end{equation}

\begin{equation}
F(a) \leq F(b), \quad a \leq b 
\end{equation}

\begin{equation}
P(a \leq x \leq b) = F(b) - F(a)
\end{equation}
